```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
#Create file path and load data files 
ad_data = "../Data/NFL_ad_spend.csv"
ad_data_df = pd.read_csv(ad_data)

ad_data_df.head(18)
```

```python
#Create a dataframe with only the neccesary information
ad_spend = ad_data_df[["Season", "Date", "Network", "Second Spot", "Viewers"]]
ad_spend.head(18)
```

```python
#plot the :30s spot video for each Superbowl from 2000-2017 
y_values = ad_spend["Second Spot"]
x_values = ad_spend["Season"]

graph = plt.subplots(figsize=(15,5))
plt.title("Cost of a :30 Second Ad Spot During the Superbowl")

plt.ylabel("Dollars (Millions)")
plt.xlabel("Year")

ad_spend["Season"] = pd.to_datetime(ad_spend["Season"], format="%Y")

plt.grid(axis="y")

#ad_spend["Second Spot"] = ad_spend["Second Spot"].str.replace("," "")
#ad_spend["Second Spot"] = ad_spend["Second Spot"].str.replace("$" "")
#ad_spend["Second Spot"] = pd.to_numeric(ad_spend["Second Spot"])

plt.plot(ad_spend["Season"], y_values)
```
