```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
merch_data = "../SB_capacity/Superbowl_Merch.csv"
merch_df = pd.read_csv(merch_data)
merch_df.head(21)
```

```python
egr = merch_df["EGR"]
y_pos = merch_df["Year"]

plt.figure(figsize=(20,10))
plt.bar(y_pos, egr, align='center', alpha=0.5)
plt.xlabel('Year')
plt.title('Economic Growth Rate')
for x,y in zip(y_pos,egr):

    label = "{0:.2f}%".format(y)

    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') 

plt.show()
```

```python
aps = merch_df["APS"]
y_pos = merch_df["Year"]

plt.figure(figsize=(20,10))
plt.bar(y_pos, aps, align='center', alpha=.5, color="red")
plt.xlabel('Year')
plt.title('Average Person Spend in Superbowl')

for x,y in zip(y_pos,aps):

    label = "${:.2f}".format(y)

    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') 


plt.show()
```
