```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
capacity_data = "../SB_capacity/Superbowl_Arena_Capacity.csv"
capacity = pd.read_csv(capacity_data)
capacity.head()
```

```python
plt.figure(figsize=(50,100))
x_values = capacity["Year"]

plt.subplot(311)
plt.title("Superbowl Attendees 1967-2018", fontsize=20)
plt.xticks(rotation=45)
plt.xlabel("Year")
plt.ylabel("Number of Attendees")
plt.plot(x_values, capacity["Attendees"], label="Attendees", color="Blue")
plt.legend()
plt.xlim(0, len(x_values))
plt.grid(axis="y")

plt.subplot(312)
plt.title("Superbowl Capacity 1967-2018", fontsize=20)
plt.xticks(rotation=45)
plt.xlabel("Year")
plt.ylabel("Arena Capacity")
plt.plot(x_values, capacity["Capacity"], label="Capacity", color="Red")
plt.legend()
plt.xlim(0, len(x_values))
plt.grid(axis="y")

 
```

```python
fig, cap = plt.subplots(1,1,figsize=(20,10))
cap.set_xlabel("Year")
cap.set_ylabel('Capacity', color='blue')
cap.plot(x_values, capacity["Capacity"], color="blue", label="Capacity")
cap.tick_params(axis='y', labelcolor="blue")


attendance = cap.twinx()  
attendance.set_ylabel('Attendees', color="red")  
attendance.plot(x_values, capacity["Attendees"], color = "red", label = "Attendance")
attendance.tick_params(axis='y', labelcolor="red")

cap.tick_params(axis='x', labelrotation=45, which='major', labelsize=15)
plt.xlim(0, len(x_values))
plt.title("Superbowl Capacity & Attendance 1967-2020", fontsize=15)
plt.tight_layout()
plt.grid()
```
