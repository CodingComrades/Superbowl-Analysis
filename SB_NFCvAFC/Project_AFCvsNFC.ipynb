```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
```

```python
#read our csv's
df = pd.read_csv("/Users/jlh/ucb-bel-data-pt-08-2020-u-c/Saved_Docs/Project_NFL/superbowl_full.csv")
df_wins = pd.read_csv("/Users/jlh/ucb-bel-data-pt-08-2020-u-c/Saved_Docs/Project_NFL/superbowl_all_games_T.csv")
df_capacity = pd.read_csv("/Users/jlh/ucb-bel-data-pt-08-2020-u-c/Saved_Docs/Project_NFL/superbowl_capacity.csv")
df_season = pd.read_csv("/Users/jlh/ucb-bel-data-pt-08-2020-u-c/Saved_Docs/Project_NFL/nfl_team_season.csv")
```

```python
#define variables
x_years = df["Season"]
y_NFC = df["NFC"]
y_AFC = df["AFC"]
y_attendance = df["Attendance"]
y_viewers = df["Viewers (MM)"]
y_rating = df["Rating"]
y_sum_wins = df["NFC-AFC Wins"]
y2_sum_wins = df["AFC-NFC Wins"]
```

```python
#first graph showing the net wins between NFC v. AFC
fig, ax = plt.subplots(figsize=(17,6))
rect1 = ax.bar(x_years, y_sum_wins, color="c", width=.95, label='NFC Net Wins')
rect2 = ax.bar(x_years, y2_sum_wins, color="m", width=.95, label="AFC Net Wins")
ax.set_ylabel('Sum of Wins')
ax.set_xlabel('Season')
ax.set_title('Interconference Net Totals, NFC v. AFC 1995-2017')
ax.legend(loc="best")
plt.savefig("Interconference_Net")
plt.xticks(x_years, rotation='vertical');
```

```python
#a comparison between rating and NFC v. AFC
fig, ax1 = plt.subplots(figsize=(16,5))
ax2 = ax1.twinx()
rating, = ax1.plot(x_years, y_rating, color='g', label="Super Bowl Rating")
nfc, = ax2.plot(x_years, y_sum_wins, color='c', label="NFC Net Wins")
fig.legend(loc="lower right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
ax1.set_xlabel('Season')
ax1.set_ylabel("Super Bowl Rating")
ax1.tick_params(axis='y', colors='green')
ax2.tick_params(axis='y', colors='c')
ax2.set_ylabel("Net Wins")
plt.title("Super Bowl Ratings vs. NFC Net Wins 1995-2017")
plt.show()
####
plt.figure(figsize=(20,4))
plt.scatter(y_rating, y_sum_wins, color='c')
plt.xlabel('Super Bowl Rating')
plt.ylabel('NFC Net Wins')
correlation_r = stats.pearsonr(y_rating, y_sum_wins)
print(f"the correlation between the Super Bowl Rating and NFC Net Wins is: {round(correlation_r[0],2)}")

rating_slope, rating_int, rating_r, rating_p, rating_std_err = stats.linregress(y_rating, y_sum_wins)
rating_fit = rating_slope * y_rating + rating_int
plt.plot(y_rating, rating_fit, "--", color="green")
plt.xticks(ticks=y_rating, rotation='vertical')
plt.show()
#####
fig, ax1 = plt.subplots(figsize=(16,5))
ax2 = ax1.twinx()
rating, = ax1.plot(x_years, y_rating, color='g', label="Super Bowl Rating")
afc, = ax2.plot(x_years, y2_sum_wins, color='m', label="AFC Net Wins")
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
ax1.set_xlabel('Season')
ax1.set_ylabel("Super Bowl Rating")
ax1.tick_params(axis='y', colors='green')
ax2.tick_params(axis='y', colors='m')
ax2.set_ylabel("Net Wins")
plt.title("Super Bowl Ratings vs. AFC Net Wins 1995-2017")
plt.show()

plt.figure(figsize=(20,4))
plt.scatter(y_rating, y2_sum_wins, color="m")
plt.xlabel('Super Bowl Rating')
plt.ylabel('AFC Net Wins')
correlation_r2 = stats.pearsonr(y_rating, y2_sum_wins)
print(f"the correlation between the Super Bowl Rating and AFC Net Wins is: {round(correlation_r2[0],2)}")

rating2_slope, rating2_int, rating2_r, rating2_p, rating2_std_err = stats.linregress(y_rating, y2_sum_wins)
rating2_fit = rating2_slope * y_rating + rating2_int
plt.plot(y_rating, rating2_fit, "--", color="green")
plt.xticks(ticks=y_rating, rotation='vertical')
plt.show()

#####
```

```python
#a comparison between AFC/NFC and viewership
fig, ax1 = plt.subplots(figsize=(16,5))
ax2 = ax1.twinx()
rating, = ax1.plot(x_years, y_viewers, color='olive', label="Super Bowl Viewers (MM)")
nfc, = ax2.plot(x_years, y_sum_wins, color='c', label="NFC Net Wins")
fig.legend(loc="lower right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
ax1.set_xlabel('Season')
ax1.set_ylabel("Super Bowl Viewers (MM)")
ax1.tick_params(axis='y', colors='olive')
ax2.tick_params(axis='y', colors='c')
ax2.set_ylabel("NFC Net Wins")
plt.title("Super Bowl Viewers (MM) vs. NFC Net Wins 1995-2017")
plt.show()

plt.figure(figsize=(20,4))
plt.scatter(y_viewers, y_sum_wins, color='c')
plt.xlabel('Super Bowl Viewers')
plt.ylabel('NFC Net Wins')
correlation_v = stats.pearsonr(y_viewers, y_sum_wins)
print(f"the correlation between NFC Net Wins and Rating is: {round(correlation_v[0],2)}")

viewers_slope, viewers_int, viewers_r, viewers_p, viewers_std_err = stats.linregress(y_viewers, y_sum_wins)
viewers_fit = viewers_slope * y_viewers + viewers_int
plt.plot(y_viewers, viewers_fit, "--", color="olive")
plt.xticks(ticks=y_viewers, rotation='vertical')
plt.show()

#####

fig, ax1 = plt.subplots(figsize=(16,5))
ax2 = ax1.twinx()
afc, = ax1.plot(x_years, y_viewers, color='olive', label="Super Bowl Viewers (MM)")
rating, = ax2.plot(x_years, y2_sum_wins, color='m', label="AFC Net Wins")
fig.legend(loc="lower right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
ax1.set_xlabel('Season')
ax1.set_ylabel("Super Bowl Viewers (MM)")
ax1.tick_params(axis='y', colors='olive')
ax2.tick_params(axis='y', colors='m')
ax2.set_ylabel("AFC Net Wins")
plt.title("Super Bowl Viewers (MM) vs. AFC Net Wins 1995-2017")
plt.show()

plt.figure(figsize=(20,4))
plt.scatter(y_viewers, y2_sum_wins, color='m')
plt.xlabel('Super Bowl Viewers (MM)')
plt.ylabel('AFC Net Wins')
correlation_v2 = stats.pearsonr(y_viewers, y2_sum_wins)
print(f"the correlation between AFC Net Wins and Super Bowl Viewers (MM) is: {round(correlation_v2[0],2)}")

viewers2_slope, viewers2_int, viewers2_r, viewers2_p, viewers2_std_err = stats.linregress(y_viewers, y2_sum_wins)
viewers2_fit = viewers2_slope * y_viewers + viewers2_int
plt.plot(y_viewers, viewers2_fit, "--", color="olive")
plt.xticks(ticks=y_viewers, rotation='vertical')
ax.tick_params(axis='y', colors='m')
plt.show()
#####
```
